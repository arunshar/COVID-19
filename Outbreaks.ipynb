```python
def get_median_buckets_dwell(merged_data,weeks,code):
    
    list_1 = []
    list_2 = []
    list_3 = []
    list_4 = []
    list_5 = []
    
    global_dic = {}
    count = 0
    for week in weeks:
        count = count + 1        
        naics_table = week.loc[merged_data['naics_code'] == code]
        dwell_list = naics_table[['bucketed_dwell_times']]['bucketed_dwell_times'].tolist()
        list_1.append(sum([dwell_time.get('<5') for dwell_time in dwell_list]))
        list_3.append(sum([dwell_time.get('21-60') for dwell_time in dwell_list]))
        list_5.append(sum([dwell_time.get('>240') for dwell_time in dwell_list]))
        if count > 39:   
            list_2.append(sum([dwell_time.get('5-10') for dwell_time in dwell_list])+sum([dwell_time.get('11-20') for dwell_time in dwell_list])) 
            list_4.append(sum([dwell_time.get('61-120') for dwell_time in dwell_list])+sum([dwell_time.get('121-240') for dwell_time in dwell_list]))
        else :
            list_2.append(sum([dwell_time.get('5-20') for dwell_time in dwell_list]))  
            list_4.append(sum([dwell_time.get('61-240') for dwell_time in dwell_list]))
        
    global_dic['<5'] = list_1
    global_dic['5-20'] = list_2
    global_dic['21-60'] = list_3
    global_dic['61-240'] = list_4
    global_dic['>240'] = list_5
    
    print (sum(global_dic['<5']))
    print (sum(global_dic['5-20']))
    print (sum(global_dic['21-60']))
    print (sum(global_dic['61-240']))
    print (sum(global_dic['>240']))
    
    return global_dic
```

```python
def get_median_buckets_dwell_for_multiplaces(merged_data,weeks,bars):
    list_1 = [0]*50
    list_2 = [0]*50
    list_3 = [0]*50
    i=0
    count=0
    global_dic = {}
    
    for week in weeks:
        count = count + 1
        for bar in bars:
            naics_table = week.loc[merged_data['safegraph_place_id'] == bar]
            dwell_list = naics_table[['bucketed_dwell_times']]['bucketed_dwell_times'].tolist()
            list_1[i] = list_1[i]+sum([dwell_time.get('21-60') for dwell_time in dwell_list])
            list_3[i] = list_3[i]+sum([dwell_time.get('>240') for dwell_time in dwell_list])
            if count > 39:   
                list_2[i] = list_2[i]+sum([dwell_time.get('61-120') for dwell_time in dwell_list])+sum([dwell_time.get('121-240') for dwell_time in dwell_list])
            else :
                list_2[i] = list_2[i]+sum([dwell_time.get('61-240') for dwell_time in dwell_list])
        i=i+1
        
    global_dic['21-60'] = list_1
    global_dic['61-240'] = list_2
    global_dic['>240'] = list_3
    return(global_dic)
```
